```python
# Drop duplicates 
data = data.drop_duplicates() 
data.nunique()
```

## Handling missing values

```python
 #calculate the percentage of missing data per every row.
for column in data.columns:
    null_count = data[column].isnull().sum()
    if null_count > 0:
        percent_nulls = (null_count / data.shape[0]) * 100
        print(f'{column} : {percent_nulls:.1f}% nulls')
        
```

```python
#Dropping high nulls columns
for column in data.columns:
    null_count = data[column].isnull().sum()
    percent_nulls = (null_count / data.shape[0]) * 100
    if percent_nulls > 1:
        data.drop(columns = column,axis = 1,inplace=True)
```

```python
data.shape
```

```python
# Fill missing values 
categorical_cols = data.select_dtypes(include=['object', 'category']).columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Impute categorical columns using mode
for col in categorical_cols:
    if col != 'DiagPeriodL90D':
        mode = data[col].mode()[0]
        data[col].fillna(mode, inplace=True)

# Impute numerical columns using mean
for col in numerical_cols:
    if col != 'DiagPeriodL90D':
        mean = data[col].mean()
        data[col].fillna(mean, inplace=True)
```
